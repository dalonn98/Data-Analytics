```python
import csv
file=open('age2023.csv',encoding='cp949')
age = csv.reader(file)


m=[]
f=[]


name = input('인구 구조를 조회 할 행정구역이나 코드 입력')

for row in age:
  if name in row[0]:
    for i in row[3:104]: # 남성 데이터(3~104열) 저장
      m.append(int(i.replace(',','')))
    for i in row[106:]: # 여성 데이터(106~207열) 저장
      f.append(int(i.replace(',','')))

print(m)
print(f)


import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.title(name + ' 지역의 인구구조')
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.show()
```

```python
import csv
file=open('age2023.csv',encoding='cp949')
age = csv.reader(file)

size = []

name = input('인구 구조를 조회 할 행정구역이나 코드 입력')

for row in age:
  if name in row[0]:
    m=0
    f=0
    for i in range(101):
      m+=int(row[i+3].replace(',',''))
      f+=int(row[i+106].replace(',',''))

size.append(m)
size.append(f)

print(size)

import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.title(name + ' 지역의 남여 성별 인구분포')
plt.pie(size, labels=['남','여'],colors=['crimson','green'],autopct='%.1f%%',startangle=90)
plt.legend()
plt.show()
```
